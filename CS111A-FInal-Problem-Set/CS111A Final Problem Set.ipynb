In other words, it represents the related rates between themain function and the sub variables that will alter the variables the function depends on.

$\frac{\partial z}{\partial v}$ = $\frac{\partial z}{\partial x}$ $\frac{dx}{dv}$ + $\frac{\partial z}{\partial y}$ $\frac{dy}{dv}$

To estimate the value of the function, we must follow the steps given below:
1. Define the relevent function first. 
2. Find the derivative of the function.
3. Now that we have both the function and its derivative, we will loop n times, adding the previous y value to the y values times the current step size multiplied with the derivative that we just calculated.
4. Plot the functionâ€™s behavior over the relevant range as well as the approximations to visually compare the estimated values found by our approximations
5. There may be some errors that come due to the approximation, which can either be computed or visualized.

```python
import numpy as np

def Eulers_solver(df, y0, n, h):
    # define initial unfilled list
    y = np.zeros(int(n)+1)
    y[0] = y0
    # iterate through steps
    for i in range(int(n)):
        # eulers solution implementation from q5
        y[i+1] = y[i] + h*df(y[i])
    return y # return value

f(x) = x^2 # define test function as x^2
df = diff(f, x)
z = Eulers_solver(df, y0=1, n=4, h=0.5) # call function with y0, n, and H values
z
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 10))
plt.plot(z)
real_ans = [1,4,9,16,25]
plt.plot(real_ans)
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Approximation for Euler's Method")
```

```python
def error(approximation, real):
    '''
    Input: 2 lists, one approximated and one real
    Output: list of relative errors for each approximation
    '''
    # make error lst
    lst = []
    # iterate through values in approx and real
    for i in range(len(approximation)):
        # append relative error to lst
        lst.append((real[i] - approximation[i]) / real[i])
    return lst # return

error(z, real_ans)
```
